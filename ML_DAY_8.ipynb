```python
x=[[180,70,8],[172,70,9],[150,50,8]]
y=['male','male','female']
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bn=BernoulliNB()
```

```python
bn_model=bn.fit(x,y)
```

```python
pr=bn_model.predict([[140,55,7]])
```

```python
pr
```

```python
pr
```
